```python
import cv2
import numpy as np
import os

# Path to class members' images
IMAGE_FOLDER = "class_portraits"

# Simulated user database
users = {
    "student1": {"matno": "12345"},
    "student2": {"matno": "67890"},
}

# Enhancement functions
def apply_histogram_equalization(image):
    return cv2.equalizeHist(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY))

def apply_gaussian_blur(image):
    return cv2.GaussianBlur(image, (5, 5), 0)

def apply_sharpening(image):
    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    return cv2.filter2D(image, -1, kernel)

def apply_edge_detection(image):
    return cv2.Canny(image, 100, 200)

# Function to process images
def enhance_image(image_path, technique):
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error loading {image_path}")
        return None

    if technique == "histogram":
        return apply_histogram_equalization(image)
    elif technique == "blur":
        return apply_gaussian_blur(image)
    elif technique == "sharpen":
        return apply_sharpening(image)
    elif technique == "edge":
        return apply_edge_detection(image)
    else:
        print("Invalid technique!")
        return None

# Function to authenticate user
def authenticate(username, matno):
    return username in users and users[username]["matno"] == matno

# Main program
username = input("Enter Username: ").strip()
matno = input("Enter Matric Number: ").strip()

if authenticate(username, matno):
    print("Login successful! Choose an enhancement technique:")
    print("1. Histogram Equalization")
    print("2. Gaussian Blur")
    print("3. Sharpening")
    print("4. Edge Detection")

    choice = input("Enter your choice (1-4): ").strip()
    techniques = {"1": "histogram", "2": "blur", "3": "sharpen", "4": "edge"}

    if choice in techniques:
        enhancement = techniques[choice]

        # Process all images in the folder
        for file in os.listdir(IMAGE_FOLDER):
            if file.endswith((".jpg", ".png")):
                img_path = os.path.join(IMAGE_FOLDER, file)
                enhanced = enhance_image(img_path, enhancement)

                if enhanced is not None:
                    cv2.imshow("Enhanced Image", enhanced)
                    cv2.waitKey(0)
                    save_path = os.path.join("enhanced_results", f"enhanced_{file}")
                    cv2.imwrite(save_path, enhanced)
                    print(f"Saved: {save_path}")

        cv2.destroyAllWindows()
    else:
        print("Invalid choice!")
else:
    print("Login failed! Invalid username or matric number.")
```
